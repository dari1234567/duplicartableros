### <u>Obtención de datos y mapeado de columnas</u>

```python
original_column_values = []

for item in  r['data']['items']:

    print(f"Item: {item}")
    
    item_name = item['name']
    column_values = item['column_values']
    print(f"column_values: {column_values}")



    for column in column_values:
        # print(column)
        temp = {}

        type = column['column']['type']
        id = column['column']['id']
        title = column['column']['title']
        # print(f"type: {type}")

        if type == 'text':
            value = column['text']

        elif type == 'date':
            value = column['date']

        elif type == 'status':
            value = column['label']
        elif type == 'people':
            value = column['persons_and_teams']
        elif type == 'dropdown':
            value = column['values']
        elif type == 'timeline':
            value = {"from": column['from'].split("T")[0], "to": column['to'].split("T")[0]}
        elif type == 'link':
            value = {"url" : column['url'], "text": column['url_text']}
        elif type == 'numbers':
            value = column['number']
        elif type == 'checkbox':
            value = {"checked": column['checked']}
        elif type == 'phone':
            value = {
            "phone": f"+{column['phone'].lstrip('+')}","countryShortName": column['country_short_name'].upper()
            }
        elif type == 'world_clock':
            value = {
                "timezone": column['timezone'] 
            }
            # print(f"ESTA COLUMNA QUE BUSCO ES LA DE TIME ZONE {value}")
            
        elif type == 'location':
            value = {
                "lat": column['lat'],
                "lng": column['lng'],
                "address": column['address']
            }
        # elif type == 'country':
        #     value = {
        #         "name": column['country']['code'],
        #         "code": column['country']['name']
        #     }
        elif type == 'country':
            # obtienes name/code desde el origen
            name = column['country']['name']
            code = column['country']['code']
            # y construyes el objeto con las claves que espera la API
            value = {
                "countryCode": code,
                "countryName": name
            }
            print(f"ESTA COLUMNA QUE BUSCO ES LA DE COUNTRY {value}")
        # elif type == 'dependency':
        #     value = {
        #         "item_ids": [column['linked_item_ids']]
        #     }
        #     print(f"ESTA COLUMNA QUE BUSCO ES LA DE DEPENDENCY {value}")
        elif type == 'email':
            value = {
                "email": column['email'],
                "text": column['text']
            }
        elif type == 'hour':
            value = {
                "hour": column['hour'],
                "minute": column['minute']
            }
        elif type == 'rating':
            value = {
                "rating": column['rating']
            }
        elif type == 'tags':
            value = {
                "tag_ids": column['tag_ids'],
            }
            print(f"ESTA COLUMNA QUE BUSCO ES LA DE TAGS {value}")
        
        elif type == 'long_text':
            value = column['text']
            
            

        else :
            value = None

        temp['tipo'] = type
        temp['value'] = value
        temp['id'] = id
        temp['title'] = title
        
        

        # print(f"temp: {temp}")
        original_column_values.append(temp)

print (f"original_column_values: {original_column_values}")
```
